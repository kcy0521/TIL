### 식별자 정의와 특징
- let : 블록 스코프 지역변수를 선언(추가로 동시에 값을 초기화)
    - 재할당 가능 & 재선언 불가능
    - 블록 스코프를 갖는 지역변수를 선언, 선언과 동시에 원하는 값으로 초기화 할 수 있음
- const : 블록 스코프 읽기 전용 상수를 선언
    - 재할당 불가능 & 재선언 불가능
    - 선언시 반드시 초기값을 설정해야 하며, 이후 값 변경이 불가능
    - let과 동일하게 블록 스코프를 가짐 
- var : 변수를 선언
    - 재할당 & 재선언 가능
    - 잘 사용하지 마라 오류 많이 난다. 


```js
let foo // 선언
foo = 11 // 할당
let bar = 0 // 선언 + 할당
console.log(bar) // 0
```

- 함수 스코프 : 함수의 중괄호 내부를 가리킴
    - 함수 스코프를 가지는 변수는 함수 바깥에서 접근 불가능

- 호이스팅  
    - var로 선언된 변수는 선언 이전에 참조할 수 있으며, 이러한 현상을 호이스팅 이라고 한다.  
    - 변수를 선언 이전에 참조할 수 있는 현상  

- const사용을 권장 / 재할당 시에만 let 사용

## 연산자

- 할당 연산자
    - +=, -=, *=, ++(증감식), --(증감식)
- 비교 연산자
    - 피연산자들을 비교하고 결과를 boolean으로 반환
    - 소문자가 대문자보다 크다
    - 알파벳 순서상 후순위가 더 크다.  

- 동등 연산자 (==)
    - 암묵적 타입변환을 통해 비교 해준다.
    - 예를 들어 a = 1, b = '1'이면 둘이 같다고 해준다.

- 일치 연산자 (===)
    - 암묵적 타입변환이 발생하지 않는다.  

- 논리 연산자
    - &&(and), ||(or), !(not)

- 삼항 연산자 
    - 3개의 피연사자를 사용하여 조건에 따라 값을 반환하는 연산자
    - 가장 앞의 족ㄴ이 참이면:(콜론) 앞의 값이 반환되며, 그반대일 경우:뒤의 값이 반환되는 연산자
    ```js
    true ? 1 : 2 // 1
    false ? 1 : 2 // 2

    const result = Math.PI > 4 ? 'Yep' : 'Nope'
    console.log(result) // Nope
    ```

## 조건문 
- if : 조건 표현식의 결과값을 boolean 타입으로 변환후 참/거짓을 판단.  
```js
if (name=== 'admin') {
    console.log('관리자 ㅎㅇ')
} else if ( name === 'manager') {
    console.log('매니저님 환영')
} else {
    console.log(`${name}님 안녕`)
}
```
    - ( 조건은 소괄화 안에 작성 )
    - 실행할 코드는 줄괄호{}안에 작성  

- switch : 조건 표현식의 결과값이 어느 값(case)에 해당하는지 판별  
```js
switch(name) {
    case '홍길동' : {
        console.log('관리자')
        break
    } case 'manager' : {
        console.log('매니저')
        break
    } default : {
        console.log(`${name} 님 하이`)
    }
}
```

- 조건이 많은 경우 switch문이 가독성이 더 좋음 / 일반적으로 중첩 else if 문은 유지보수가 힘들다는 문제도 있음


## 반복문 

- while : 조건문이 참이기만 하면 문장을 계속해서 수행
- for : 특정한 조건이 거짓으로 판별될 때까지 반복
- for .. in : 객체의 속성을 순회할 때 사용
- for .. of : 반복 가능한 객체를 순회할 때 사용
    - 반복가능 객체의 종류 : Array, Set, String 등

- for ..in 과 for .. of의 차이 
    - 속성이름 (인덱스값을 나타냄,key값 나타냄) / 속성 값 (value값을 나타냄)
