일반적인 아이디어 도출  
- 회사 : 상급자의 결재가 난 의견  
- 프로젝트 수주: 고객의 요청에 의한 프젝  
- 명확한 목표가 정해진 상태로 요구사항 정의를 들어가기때문에 초반 어려움이 적음  

창의적인 아이디어 도출  
- 기존경험 없고, 있으면 좋고, 우리가 할 수 있을만 한거  
- 우리가 생각한 대부분은 이미 시중에 존재  
- 동등한 위치에서 회의 결정권자 없음(회의시간 증가)  

기획 팁  
- 명세서 기반 프젝 큰틀 구체화 하기  
- 시간정해두고하기, 일정연장 하지 말기  
- 너무 새로운것 집착 노, 기존 개선방식도 고려해보기  

프로젝트 규모(구현 가능성 & 기술력, 프젝 차별성)
- 각자의 기술스택 고민
- 핵심기능 명확하게 도출, 개발시 우선되어야함
- 시간부족, 필요없는것 버리기 과감필요
- 포트폴리오에 쓰기엔 싸피 졸업생들 많아서 규모 짐작 하기 쉬움
- 혼자하는것 보단 큰 프젝 차별성 두기

프젝 차별성
- 내가 당장 필요한 사이트 만들라고함
- 가장 중요한 부분기획 구체화 잘함
- 사이트 활성화 이후의 것만 생각함
- 행복회로 돌리지말고 부정적인면 보기
- 최초 유저를 끌어들이는 방법, 유저가 관심 있어 할 기능에 대한 고민
- 유저 설득 당위성(돈, 페러다임)
- 포괄적인 기능을 가지고 있는 것들은 정말 개발하기 어렵다. (블로그 등)
- 명확한 주제가 있는 프젝하는 것이 개발 범위와 일정 맞출 수 있다.
- `6주만에 노션보다 좋은 블로그 만들 수 있는지 고민`
- 우리 프젝 핵심은 무엇인지 고민 우리의 장점은 무엇인가?
